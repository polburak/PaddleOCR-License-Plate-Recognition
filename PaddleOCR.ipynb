#KOD BLOĞU (Çoklu Kontrol ve Klasöre Kayıt Etme)
 eğitilen model seti ile verilen plakaların kontrolü yaparak farklı bir klasöre aktarıp üzerine OCR ile plakayı yazdırır.

```python
import glob, os, cv2, numpy as np
from paddleocr import PaddleOCR

ocr = PaddleOCR(
    use_angle_cls=True,
    text_detection_model_dir='/content/ch_ppocr_server_v2.0_det_infer',
    text_recognition_model_dir='/content/rec_plaka_inference'
)

image_folder = "/content"
output_folder = "/content/ocr_results"
os.makedirs(output_folder, exist_ok=True)

image_paths = glob.glob(os.path.join(image_folder, "*.jpg")) + \
              glob.glob(os.path.join(image_folder, "*.png"))

for img_path in image_paths:
    print(f"\n--- İşleniyor: {os.path.basename(img_path)} ---")
    result = ocr.ocr(img_path)
    img = cv2.imread(img_path)

    if not result or not result[0]:
        print("⚠️ Plaka bulunamadı.")
    else:
        for line in result[0]:
            box = line[0]
            text, score = line[1]

            # Kutuyu çiz
            pts = [(int(x), int(y)) for x, y in box]
            cv2.polylines(img, [np.array(pts)], isClosed=True, color=(0,255,0), thickness=2)

            # Yanına yazı ekle
            x, y = pts[0]
            cv2.putText(img, f"{text} ({score:.2f})", (x, y-10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0,0,255), 2)

            print(f"Plaka: {text} (confidence: {score:.2f})")

    out_path = os.path.join(output_folder, os.path.basename(img_path))
    cv2.imwrite(out_path, img)
    print(f"✅ Kaydedildi: {out_path}")
```
